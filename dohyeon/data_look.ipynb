```python
import pandas as pd


raw_train = pd.read_csv('train.csv', encoding = 'utf-8')
```

```python
raw_train
```

```python
import ast
raw_train['intent'] = 0
```

```python
import ast
raw_train['intent'] = 0
for index in range(len(raw_train['intent'])):
    raw_train['intent'][index] = ast.literal_eval(raw_train['dataSet.answer.intent'][index])[0]['category']
```

```python
raw_train
```

```python
raw_train['case'] = 0
```

```python
for index in range(len(raw_train)):
    if raw_train['intent'][index] in ["technology","attitude"]:
        raw_train['case'][index] = 'job'
    elif raw_train['intent'][index] in ["background", "personality" ,"etc"]:
        raw_train['case'][index] = 'personality'
```

```python
raw_train['case'].unique() # 혹시나 누락된부분 있나 확인차.
```

```python
raw_train['age']= 0
```

```python
raw_train['dataSet.info.ageRange'].value_counts()
```

```python
# 각 요소에서 '-' 문자 제거 후 'age' 컬럼에 저장
for index in range(len(raw_train)):
    # '-' 제거
    cleaned_value = raw_train.loc[index, 'dataSet.info.ageRange'].replace('-', '')
    # 'age' 컬럼에 저장
    raw_train.loc[index, 'age'] = int(cleaned_value[-2:])

    
```

```python
raw_train['age'].value_counts()
```

```python
use_train = raw_train[['dataSet.info.occupation', 'dataSet.info.place', 'dataSet.info.gender','dataSet.info.experience', 'dataSet.question.raw.text','dataSet.answer.raw.text', 'dataSet.answer.summary.text', 'case','age']]

```

```python
use_train.to_csv('use_train_data.csv', index=False, encoding='utf-8')
```
